## Task 1

Implementation of some basic vector and matrix calculations using NumPy.

Reference: https://numpy.org/doc/1.18/reference/index.html

Create the following vectors and matrix using numpy. They should use floating point as data type.
$$
a = \begin{bmatrix}
1 & 0 & 1 & 0 & 1
\end{bmatrix}
$$

$$
b = \begin{bmatrix}
1 & 2 & 3 & 4 & 5
\end{bmatrix}
$$

$$
M =\begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6\\
7 & 8 & 9\\
\end{bmatrix}
$$

Also define the following scalar value
$$k = 5.0$$

````python
# TODO: Create vectors, matrix and scalar value

# Create vectors
a = np.array([1, 0, 1, 0, 1]) 
b = np.arange(1, 6, 1)

# Create a matrix
M = np.matrix(np.arange(1, 10, 1).reshape((3, 3)))

# Create a scalar value
k = np.float16("5.0")

print(f'a = {a}')
print(f'b = {b}')
print(f'M = {M}')
print(f'k = {k}')
````

````python
# TODO: Add vector a with vector b
add = a + b
np.testing.assert_allclose(add, np.array([2, 2, 4, 4, 6]))
print(f'a + b = {add}')

# TODO: Multiply vector a with scalar value k
mult = a * k
np.testing.assert_allclose(mult, np.array([5, 0, 5, 0, 5]))
print(f'k * a = {mult}')

# TODO: Dot product between a and b
dot = np.dot(a, b)
np.testing.assert_allclose(dot, 9.0)
print(f'<a, b> = {dot}')

# TODO: Create vector c by sclicing vector a in order to get [0, 1, 0]
c = a[1:-1]
np.testing.assert_allclose(c, [0, 1, 0])
print(f'c = {c}')

# TODO: Multiply vector c with matrix M
cM_mult = np.multiply(c, M)
np.testing.assert_allclose(cM_mult, [[0, 2, 0], [0, 5, 0], [0, 8, 0]])
print(f'c * M = {cM_mult}')

# TODO: Dot product between vector and matrix M
cM_dot = np.asarray(c.dot(M))[0]
np.testing.assert_allclose(cM_dot, [4, 5, 6])
print(f'<c, M> = {cM_dot}')

# TODO: Use a numpy-herlper function to create a vector d = [0, 1, 2 ..., 99]
d = np.arange(0, 100, 1)

# TODO: Now use slicing with strides to reduce this vector to d_small = [40, 50, 60]
d_small = d.reshape(10, 10)[4:7, 0]
np.testing.assert_allclose(d_small, [40, 50, 60])

print(f'd_small = {d_small}')
````

## Task 3

Plot the color histogram of the rgb image.

Reference: https://matplotlib.org/users/index.html

````python
def plot_color_histogram(image):
    '''
    Plot the histogram of each color channel into a single plot.
    Take a look at the scikit-image submodule skimage.exposure

    image -- RGB image
    '''
    # TODO: implement this function
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    ax = axes.ravel()
    
    # Set channel_axis for multichannel image
    hist, hist_centers = ski.exposure.histogram(image, nbins=256, source_range="image", channel_axis=-1)
    # Extract RGB channel
    r = hist[0, :]
    g = hist[1, :]
    b = hist[2, :]

    ax[0].imshow(image)
    ax[0].set_title("Input Image")

    ax[1].plot(hist_centers, r, color='red', label='Red')
    ax[1].plot(hist_centers, g, color='green', label='Green')
    ax[1].plot(hist_centers, b, color='blue', label='Blue')

    ax[1].set_title("Histogram of Multichannel Image")
    ax[1].set_xlabel("Value")
    ax[1].set_ylabel("Occurrence")
    ax[1].legend()

    fig.tight_layout()
    plt.show()

plot_color_histogram(image)
````

## Task 4

Convert the RGB image into the HSV color space. Rotate the colors in the HSV color space, convert it back to the RGB color space and plot it together with its histogram. For this task, you are only allowed to use NumPy.

More information about the HSV color space: https://en.wikipedia.org/wiki/HSL_and_HSV

````python
def rgb2hsv(image):
    '''
    Convert the rgb image into hsv color space and return it.

    image -- rgb image
    '''
    # TODO: implement function

    if image.shape[-1] != 3:
        raise ValueError("Image must have 3 channels (RGB).")

    # Step 1: Compute the maximum and minimum component (indices and values) in RGB space (operation along channel axis)
    max_index = np.argmax(image, axis=-1)
    x_max = np.max(image, axis=-1)
    x_min = np.min(image, axis=-1)
    c = x_max - x_min      # chroma
    l = (x_max + x_min)/2  # luminance

    # Step 2: Create HSV space
    v = x_max                          # value
    s = np.where(v != 0, c/v, 0)       # saturation
    h = np.zeros_like(v)               # hue

    # Extract RGB
    r = image[:,:,0]
    g = image[:,:,1]
    b = image[:,:,2]

    # Case 1: Red is max
    mask_r = (v == r) & (c != 0)
    h[mask_r] = 60 * (0 + (g[mask_r] - b[mask_r]) / c[mask_r])

    # Case 2: Green is max
    mask_g = (v == g) & (c != 0)
    h[mask_g] = 60 * (2 + (b[mask_g] - r[mask_g]) / c[mask_g])

    # Case 3: Blue is max
    mask_b = (v == b) & (c != 0)
    h[mask_b] = 60 * (4 + (r[mask_b] - g[mask_b]) / c[mask_b])

    # Step 3: Stack HSV
    hsv = np.stack((h, s, v), axis=-1)
    return hsv
````

````python
def hsv2rgb(image):
    '''
    Convert the hsv image into rgb color space and return it.

    image -- rgb image
    '''
    # TODO: implement function

    if image.shape[-1] != 3:
        raise ValueError("Image must have 3 channels (HSV).")

    # Extract HSV
    h, s, v = image[:, :, 0], image[:, :, 1], image[:, :, 2]

    # Step 1: Compute chroma and divide the 360 spectrum into its components
    c = v * s
    h = h / 60
    x = c * (1 - np.abs(np.mod(h, 2) -1))

    # Step 2: Compute pure RGB components
    rgb = np.zeros_like(image)

    mask = (h >= 0) & (h < 1)
    rgb[mask, 0] = c[mask]
    rgb[mask, 1] = x[mask]
    
    mask = (h >= 1) & (h < 2)
    rgb[mask, 0] = x[mask]
    rgb[mask, 1] = c[mask]
    
    mask = (h >= 2) & (h < 3)
    rgb[mask, 1] = c[mask]
    rgb[mask, 2] = x[mask]
    
    mask = (h >= 3) & (h < 4)
    rgb[mask, 1] = x[mask]
    rgb[mask, 2] = c[mask]
    
    mask = (h >= 4) & (h < 5)
    rgb[mask, 0] = x[mask]
    rgb[mask, 2] = c[mask]
    
    mask = (h >= 5) & (h < 6)
    rgb[mask, 0] = c[mask]
    rgb[mask, 2] = x[mask]
    
    # Add the white proportion
    m = v - c
    rgb = rgb + m[..., np.newaxis]
    
    return rgb.astype(np.uint8)
````

````python
# Change the color space from rgb to hsv and plot it
image_hsv = rgb2hsv(image)

# TODO: plot image
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
ax[0].imshow(image_hsv)
ax[0].set_title("HSV")
ax[1].imshow(image_hsv[...,0]/360)
ax[1].set_title("HSV with Normalized Hue")

# TODO: Rotate the colors in hsv color space
rotation_angle = 180  # degrees
image_rotation = image_hsv.copy()
image_rotation[..., 0] = (image_rotation[..., 0] + rotation_angle) % 360


# TODO: Now, convert it back from HSV to RGB and plot it
image_rgb = hsv2rgb(image_rotation)
ax[2].imshow(image_rgb)
ax[2].set_title(f"Image after rotation of ({rotation_angle} degrees)")

# TODO: Plot the color histogram of the inverted rgb image
plot_color_histogram(image_rgb)
````
